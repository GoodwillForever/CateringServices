### 任务1.2 处理菜品名称字符数据

```python
meal_detail["dishes_name"] = meal_detail["dishes_name"].str.replace("\r\n","") # 使用 replace方法将 其中的空白字符转化成 ""
meal_detail["dishes_name"].head()
```

### 任务1.3 构建热销度评分指标(使用Min-Max标准化)

```python
foodName_sort = meal_detail.groupby(by="dishes_name")["counts"].sum().sort_values(ascending=False) # 因为未来做的是菜品推荐系统 所以 白饭/小碗 和白饭/大碗 可以去除
foodName_sort = foodName_sort[2:]
foodName_sort.head()
```

```python
min_max = (foodName_sort - foodName_sort.min())/(foodName_sort.max() - foodName_sort.min())
min_max
```

### 任务1.4 绘制条形图展示热销榜排名Top10

```python
top10=min_max[:10]
plt.figure(figsize=(20,8),dpi=80)
top10.plot(kind='barh', color='b',alpha=0.4)
plt.xticks(fontsize=12)
plt.yticks(fontproperties='SimHei',fontsize=12)
plt.xlabel("")
plt.ylabel("")
plt.title('热销榜菜品Top10',fontproperties='SimHei',fontsize=16)
plt.show()
```
